```python
spotify_songs = len(spotify)
tiktok_songs = len(tiktok)

spotify_artist_counts = spotify['artist_name'].value_counts().reset_index()
spotify_artist_counts.columns = ['artist_name', 'count']
spotify_artist_counts['data_src'] = 'spotify'
spotify_artist_counts['sizeratio'] = spotify_artist_counts['count'] / spotify_songs

tiktok_artist_counts = tiktok['artist_name'].value_counts().reset_index()
tiktok_artist_counts.columns = ['artist_name', 'count']
tiktok_artist_counts['data_src'] = 'tiktok'
tiktok_artist_counts['sizeratio'] = tiktok_artist_counts['count'] / tiktok_songs
```

```python
len(tiktok_artist_counts), len(spotify_artist_counts)
```

```python
common_artists = set(spotify_artist_counts['artist_name']).intersection(set(tiktok_artist_counts['artist_name']))

common_artist_count = []
for artist in common_artists:
    spotify_count = spotify_artist_counts.loc[spotify_artist_counts['artist_name'] == artist, 'count'].iloc[0]
    tiktok_count = tiktok_artist_counts.loc[tiktok_artist_counts['artist_name'] == artist, 'count'].iloc[0]
    sizeratio = spotify_artist_counts.loc[spotify_artist_counts['artist_name'] == artist, 'sizeratio'].iloc[0]
    common_artist_count.append({'artist_name': artist, 'spotify_count': spotify_count, 'tiktok_count': tiktok_count, 'sizeratio':sizeratio, 'data_src':'both'})

common_artist_count = pd.DataFrame(common_artist_count)

spotify_artist_counts = spotify_artist_counts[~spotify_artist_counts['artist_name'].isin(common_artists)]
tiktok_artist_counts = tiktok_artist_counts[~tiktok_artist_counts['artist_name'].isin(common_artists)]
```

```python
len(common_artist_count), len(tiktok_artist_counts), len(spotify_artist_counts)
```

```python
common_artist_count.head()
```

```python
tiktok_artist_counts.head()
```

```python
combined_artists = pd.merge(spotify_artist_counts, tiktok_artist_counts, on='artist_name', how='outer', suffixes=('_spotify', '_tiktok'))
combined_artists = pd.merge(combined_artists, common_artist_count, on='artist_name', how='outer', suffixes=('', '_other'))

combined_artists['spotify_count'].fillna(0, inplace=True)
combined_artists['tiktok_count'].fillna(0, inplace=True)
```

```python
combined_artists.head(-5)
```

```python
sizeratio_columns = ['sizeratio_spotify', 'sizeratio_tiktok', 'sizeratio']

for column in sizeratio_columns:
    combined_artists[column].fillna(0, inplace=True)

combined_artists['sizeratio'] = combined_artists[sizeratio_columns].apply(lambda row: next((cell for cell in row if cell != 0), 0), axis=1)

combined_artists.drop(columns=['sizeratio_spotify', 'sizeratio_tiktok'], inplace=True)
```

```python
data_src_columns = ['data_src_spotify', 'data_src_tiktok', 'data_src']

for column in data_src_columns:
    combined_artists[column].fillna('', inplace=True)

combined_artists['data_src'] = combined_artists[data_src_columns].apply(lambda row: next((cell for cell in row if cell != ''), ''), axis=1)

combined_artists.drop(columns=['data_src_spotify', 'data_src_tiktok'], inplace=True)
```
